```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Generate synthetic data
np.random.seed(42)
n_samples = 1000

# Create a DataFrame
data = pd.DataFrame({
    'accident_id': range(n_samples),
    'date': pd.date_range(start='2023-01-01', periods=n_samples, freq='H'),
    'road_condition': np.random.choice(['dry', 'wet', 'icy', 'foggy'], size=n_samples),
    'weather': np.random.choice(['clear', 'rain', 'snow', 'fog'], size=n_samples),
    'time_of_day': pd.date_range(start='2023-01-01', periods=n_samples, freq='H').hour,
    'latitude': np.random.uniform(low=-90.0, high=90.0, size=n_samples),  # Random latitude
    'longitude': np.random.uniform(low=-180.0, high=180.0, size=n_samples)  # Random longitude
})

# Display the first few rows
print(data.head())
```

```python
# Count accidents by road condition
road_condition_counts = data['road_condition'].value_counts()

# Count accidents by weather
weather_counts = data['weather'].value_counts()

# Count accidents by time of day
time_of_day_counts = data['time_of_day'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=road_condition_counts.index, y=road_condition_counts.values)
plt.title('Accidents by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Number of Accidents')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=weather_counts.index, y=weather_counts.values)
plt.title('Accidents by Weather')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=time_of_day_counts.index, y=time_of_day_counts.values)
plt.title('Accidents by Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# Create a scatter plot to visualize accident hotspots
plt.figure(figsize=(10, 10))
plt.scatter(data['longitude'], data['latitude'], alpha=0.5)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)  # Optional: Add grid for better visibility
plt.show()
```
